## Exploratory Data Analysis

#### Exploratory Data Analysis Tests Starts

### Q.1.For 'ABCD 2', what is the Show id and Who is the Director of the show ?

```python
#to check particular record
df[df['Title'].isin(["ABCD 2"])]
```

```python
df[df['Title'].str.contains("ABCD 2")]
```

### Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
#counting maximum shows in year
df['Date_N']=pd.to_datetime(df['Release_Date'])
```

```python
df.head()
```

```python
df['Date_N'].dt.year.value_counts()
```

```python
#Representation of tv shows and movies in a graph
df['Date_N'].dt.year.value_counts().plot(kind='bar')
plt.title("Highest number of the TV shows & Movies were released in a year")
```

### Q. 3) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
#Grouping Unique Item and Counting their values
df.groupby('Category').Category.count()
```

```python
#Representing the grouped items in bar graph
df.groupby('Category').Category.count().plot(kind='bar')
plt.title("Number Of Movies And TV Shows")
```

### Q.4.Show all the Movies that were released in year 2020.

```python
#Creating Column for years
df['Year']=df['Date_N'].dt.year
```

```python
# representing for all movies data released in 2020
df[(df['Category']=='Movie') & (df['Year']==2020)]
```

### Q.5.Show the all TV Show Titles That has been released in India

```python
# representing tvshows titles which has been released by india
df[(df['Category']=='TV Show')&(df['Country']=="India")]['Title']
```

### Q.6.Top 10 Directors contributed towards netflix

```python
# Most No films contributed by directors to netflix
df['Director'].value_counts().head(10)
```

###  Q.7.Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".

```python
# Representing the data frame as per requirement
df[((df['Category']=="Movie") & (df['Type']=="Comedies")) | (df["Country"]=="United Kingdom")]
```

### Q. 9.1) How many Movies got the 'TV-14' rating, in Canada ?

```python
# Getting the no of movies with TV-14 Rating
df[(df['Category']=="Movie") & (df['Rating']=='TV-14') & (df['Country']=="Canada")].shape
```

### Q. 9.2) How many TV Shows got the 'R' rating, after year 2018 ?

```python
# Getting the no of TV Shows with R rating after 2018
df[(df['Category']=="TV Show") & (df['Rating']=="R") & (df['Year']>2018)].shape
```

### Q. 10) What is the maximum duration of a Movie/Show on Netflix ?

```python
df.head()
```

```python
# getting the unique duration values
df['Duration'].unique()
```

```python
# checking the duration data type
df.Duration.dtype
```

```python
#Splitting up the duration field into minutes and unit as there some values in seasons units
df[['Minutes','Unit']]=df['Duration'].str.split(' ',expand=True)
```

```python
# maximum value of duration
df['Minutes'] = df['Minutes'].astype(int)
df['Minutes'].max()
```

### Q. 11) Which individual country has the Highest No. of TV Shows ?

```python
# Reprsenting the country with highest no of tv shows
tvshows=df[df['Category']=="TV Show"]
tvshows.Country.value_counts().head(1)
#automatically sorted here
```

### Q. 12) How can we sort the dataset by Year ?

```python
# Sorting dataset in descending order
df.sort_values(by="Year",ascending=False).head()
```

### Q. 13) Find all the instances where: Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids' TV'.

```python
#Representing data frame as per the instance
df[((df['Category']=="Movie")&(df['Type']=='Dramas')) | ((df['Category']=="TV Show")&(df['Type']=="Kids' TV"))].head()
```
